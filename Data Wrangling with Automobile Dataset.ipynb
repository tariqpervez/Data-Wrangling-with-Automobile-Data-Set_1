<h1>Data Wrangling of Automobile Data Set<h1> 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('imports-85.data', header=None)
df.head()
```

```python
headers=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
#inserting header titles into columns
df.columns=headers
df.head()
```

Identifying & Handling Missing Values

```python
#identifying and handling missing values
import numpy as np
df.replace('?', np.nan, inplace=True)
df.head()
```

```python
#Evaluating for missing data
missing_data=df.isnull()
missing_data.head()
```

```python
#count missing values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#determining datatypes
print(df.dtypes)
```

Dealing with missing data



1.Normalized Losses

```python
avg_norm_losses=df['normalized-losses'].astype('float').mean(axis=0)
print("Average normalized losses:", avg_norm_losses)
```

```python
#replacing missing values with averages 
df['normalized-losses'].replace(np.nan, avg_norm_losses, inplace=True)
df.head()
```

2. Bore

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average bore:", avg_bore)
```

```python
#replacing bore missing values with average bore value
df['bore'].replace(np.nan, avg_bore, inplace=True)
df[['bore']]
```

3. Stroke

```python
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average stroke:", avg_stroke)
```

```python
#replacing stroke missing values with average stroke value
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

4.Horsepower

```python
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
#replacing horsepower missing values with average horsepower value
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

5.Peak-RPM

```python
avg_peak_rpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak-rpm:", avg_peak_rpm)
```

```python
#replacing peak-rpm missing values with average peak-rpm value
df['peak-rpm'].replace(np.nan, avg_peak_rpm, inplace=True)
```

```python
#determining count of cars with respect to number of doors
df['num-of-doors'].value_counts()
```

```python
print("Automobile Count with maximum number of doors:", df['num-of-doors'].value_counts().idxmax())
```

```python
#replacing num-of-doors missing values with four 
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
#dropping rows that do not have price 
df.dropna(subset=['price'], axis=0, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
#checking data types
df.dtypes
```

```python
#converting datatypes to proper format
df["bore"].astype('float')
df["stroke"].astype('float')
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

Data Standardization 

```python
df.head()
```

Conversion unit L/100km = 235 / mpg

```python
df['city-mpg']=235/df['city-mpg']
df.head()
```

```python
#renaming city-mpg to city-L/100km
df.rename(columns={'city-mpg':'city-L/100km'}, inplace=True)
df.head()
```

```python
#converting highway-mpg to highway-L/100km
df['highway-mpg']=235/df['highway-mpg']
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df.head()
```

Data Normalization

```python
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
```

```python
df[['length','width','height']].head()
```

Binning

```python
df.dtypes
```

```python
df['horsepower']=df['horsepower'].astype('int', copy=True)
```

```python
bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
group_name=['high','medium','low']
df['horsepower-binned']=pd.cut(df['horsepower'], bins, labels=group_name,include_lowest=True)
df['horsepower-binned'].head(20)
```

```python
df['horsepower-binned'].value_counts()
```

```python
#plotting distribution of each bin 
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_name, df['horsepower-binned'].value_counts())

plt.pyplot.xlabel('Horsepower')
plt.pyplot.ylabel('Count')
plt.pyplot.title('Horsepower Bins')
```

```python
#generating histogram plot to visualizing the bin 
plt.pyplot.hist(df["horsepower"], bins = 3)
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

Dummy Variables 

```python
#converting categorical variables to dummy variables
df.head()
```

```python
dummy_variable_1=pd.get_dummies(df['fuel-type'])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas','fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
#combining dummy variable column to main dataframe
df=pd.concat([df, dummy_variable_1], axis=1)
df.drop('fuel-type', axis=1, inplace=True)
df.head()
```

```python
dummy_variable_2=pd.get_dummies(df['aspiration'])
dummy_variable_2.head()
```

```python
dummy_variable_2.rename(columns={'std':'aspiration-std','turbo':'aspiration-turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
df=pd.concat([df, dummy_variable_2], axis=1)
df.drop('aspiration', axis=1, inplace=True)
df.head()
```
